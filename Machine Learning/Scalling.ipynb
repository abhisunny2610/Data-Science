```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler, QuantileTransformer, PowerTransformer
```

```python
df = pd.read_csv("../Dataset/housing.csv")
df.head()

df2 = df.drop(columns=["ocean_proximity"])
```

```python
scaler = MinMaxScaler()
scaler.set_output(transform="pandas")
scaler_value =scaler.fit_transform(df["median_house_value"].to_numpy().reshape(-1,1))
scaler_value.describe()
```

```python
maxAbs = MaxAbsScaler()
maxAbs.set_output(transform="pandas")
maxAbs_data = maxAbs.fit_transform(df["median_house_value"].to_numpy().reshape(-1,1))
maxAbs_data.describe()
```

```python
quantile = QuantileTransformer()
quantile.set_output(transform="pandas")
quantile_data =quantile.fit_transform(df.drop(columns=["ocean_proximity"]))
```

```python
power = PowerTransformer()
power.set_output(transform="pandas")
power_data = power.fit_transform(df.drop(columns=["ocean_proximity"]))
power_data.describe()
```

```python
quantile_data.plot.hist()
```

```python
fig, axes = plt.subplots(figsize=(16,2*15), nrows=9, ncols=2)

for col, (ax1, ax2) in zip(df2.columns ,axes):
    ax1.hist(df2[col])
    ax2.hist(quantile_data[col], bins=40)
```

```python
fig, axes = plt.subplots(figsize=(16,2*15), nrows=9, ncols=2)

for col, (ax1, ax2) in zip(df2.columns ,axes):
    ax1.hist(df2[col], color='red')
    ax2.hist(power_data[col], bins=40)
```
