```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("../Dataset/bike_buyers_clean.csv")
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
new_df = encoder.fit_transform(df[["Gender", "Home Owner"]])
```

```python
feature_out = pd.DataFrame(new_df, columns=encoder.get_feature_names_out())
```

```python
names_in = encoder.feature_names_in_
```

```python
df1= df.drop(columns=names_in)
```

```python
df2 = pd.concat([df1, feature_out], axis=1)
```
