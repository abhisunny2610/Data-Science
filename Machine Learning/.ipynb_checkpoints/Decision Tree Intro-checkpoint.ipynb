```python

# Plotting
fig, axs = plt.subplots(len(df.columns), 1, figsize=(8, len(df.columns) * 4))

for i, column in enumerate(df.columns):
    # Skip non-numeric columns and the target column
    if df[column].dtype not in [int, float]:
        continue

    # Calculate probabilities
    conditions = [
        (f"{column} > {threshold}", df[df[column] > threshold]["Purchased Bike"])
        for threshold in df[column].unique()
    ]

    # Plotting
    axs[i].bar(["No Purchase", "Purchase"], [1 - (data == "Yes").mean() for _, data in conditions], color=['red', 'green'])
    axs[i].set_title(f"Probability for {column}")

plt.tight_layout()
plt.show()
```

```python

# Plotting
fig, axs = plt.subplots(len(df.columns) - 1, 1, figsize=(8, (len(df.columns) - 1) * 4))

for i, column in enumerate(df.columns):
    # Skip the target column
    if column == "Purchased Bike" and column == "ID":
        continue

    # Calculate probabilities for each unique value in the column
    conditions = [
        (f"{column} == '{value}'", df[df[column] == value]["Purchased Bike"])
        for value in df[column].unique()
    ]

    # Plotting
    bar_labels = ["No Purchase", "Purchase"]
    heights = [1 - (data == "Yes").mean() for _, data in conditions]

    # Ensure consistent lengths
    if len(bar_labels) == len(heights):
        axs[i].bar(bar_labels, heights, color=['red', 'green'])
        axs[i].set_title(f"Probability for {column}")
    else:
        print(f"Skipping {column} due to shape mismatch")

plt.tight_layout()
plt.show()
```

```python
# Plotting
fig, axs = plt.subplots(len(df.columns) - 1, 1, figsize=(8, (len(df.columns) - 1) * 4))

for i, column in enumerate(df.columns):
    # Skip the target column
    if column == "Purchased Bike":
        continue

    # Calculate probabilities for each unique value in the column
    conditions = [
        (f"{column} == '{value}'", df[df[column] == value]["Purchased Bike"])
        for value in df[column].unique()
    ]

    # Plotting
    heights = [1 - (data == "Yes").mean() for _, data in conditions]
    axs[i].bar(["No Purchase", "Purchase"], heights, color=['red', 'green'])
    axs[i].set_title(f"Probability for {column}")

plt.tight_layout()
plt.show()
```

```python
group = df.groupby("Education")
summary = group.apply(lambda x: x.groupby("Purchased Bike").count()["ID"])
summary["Proportion"] = summary["Yes"] / summary.sum(axis=1)
summary
```

```python
sample_summary = df["Purchased Bike"].value_counts(normalize=True)

group = df.groupby("Education")
summary = group.apply(lambda x: x.groupby("Purchased Bike").count()["ID"])
summary["Proportion"] = summary["Yes"] / summary.sum(axis=1)
difference = sample_summary["Yes"] - (summary["Proportion"].sum() / len(summary["Proportion"]))
difference
```

```python
difference_list = []

cols = df.iloc[:,:-1].columns

for col in cols:
    group = df.groupby(col)
    summary = group.apply(lambda x: x.groupby("Purchased Bike").count()[col])
    summary["Proportion"] = summary["Yes"] / summary.sum(axis=1)
    difference = sample_summary["Yes"] - (summary["Proportion"].sum() / len(summary["Proportion"]))
    difference_list.append(difference)
```
