```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Bar Graph

```python
fruits = ["Banana", "Graphs", "Orange", "Papaya"]
counts = [4,10,7,5]

plt.bar(x=fruits, height=counts)
plt.show()
```

## Bar color

```python
fruits = ["Banana", "Graphs", "Orange", "cherry"]
counts = [4,10,7,5]
bar_colors = ["yellow", "green", "orange", "red"]

plt.bar(x=fruits, height=counts, color=bar_colors)
plt.show()
```

## Chart labels and legends and Title

```python
fruits = ["Banana", "Graphs", "Orange", "Papaya"]
counts = [4,10,7,5]
bar_colors = ["yellow", "green", "orange", "red"]
bar_labels = ["Banana", "Graphs", "Orange", "Cherry"]

## create a figure
fig, axes = plt.subplots()

## create bars
p = axes.bar(x=fruits, height=counts, color=bar_colors, label = bar_labels)

## set x label
axes.set_xlabel("Fruits")

## set y label
axes.set_ylabel("Counts")

## set legends
axes.legend()

## set bar labels
axes.bar_label(p, label_type='center')

## set fig title
plt.title("Fruits and their counts")

## show figure
plt.show()
```

## Grouped bar, stacked bar and horizontal bar

```python
religion = ["Hinduism", 'Islam', "Sikhism", 'Christianity']
sex_counts = {
    'Male': np.array([60, 40, 50, 70]),
    'Female': np.array([40,60,50,30])
}

# df_religion = pd.DataFrame(data=sex_counts, index=religion)

fig, ax = plt.subplots(figsize=(6,4), tight_layout=True)
width = 0.6

bottom = np.zeros(4)

for sex, sex_count in sex_counts.items():
    p = ax.bar(religion, sex_count, width=width, bottom=bottom, label=sex)
    bottom += sex_count
    ax.bar_label(p, label_type="center")

ax.set_title("Number of Religion by sex")
ax.legend(loc="upper right", ncol=2)
plt.show()
```

```python
## horizontal bar chart

people = ["Hanish", "Manish", "Abhi", "Saurabh", "Sumit"]
performance = np.random.randint(8, 15, 5)
set_label = np.arange(len(people))
fig, axes = plt.subplots()

hbars = axes.barh(set_label, performance,xerr=1, align='center', color='r')
axes.bar_label(hbars)
axes.set_yticks(set_label, labels=people)
axes.set_xlabel("Performance")
axes.set_ylabel("Name")
ax.set_xlim(right=15)

plt.show()
```

## Group bar chart with labels

```python
species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

labels = np.arange(len(species))

fig, ax = plt.subplots()

width=0.3
offset = 0

for col, values in penguin_means.items():
    rgb = (random.random(), random.random(), random.random())
    bars = ax.bar(labels+offset, values, width=width, label=col, color=[rgb])
    offset +=width
    ax.bar_label(bars, padding=3)

ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(labels + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0,250)
plt.show()
```
