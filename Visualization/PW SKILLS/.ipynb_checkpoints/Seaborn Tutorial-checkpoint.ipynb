# Seaborn Tutorial

### Seaborn: statistical data visualization¶
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Line Chart

```python
# plotting lineplot using sns.lineplot()
plt.style.use('seaborn-darkgrid')
%matplotlib inline
tips.sort_values("total_bill")
sns.lineplot(x=tips["total_bill"], y=tips["tip"])
plt.show()
```

```python
"""   - Adjusting background color using axes.facecolor
      - Changing label size using xtick.labelsize, ytick.labelsize
"""

plt.figure(figsize=(14,10))
sns.set(rc={"axes.facecolor": "#FFFDE7", "axes.grid": True, "xtick.labelsize": 14, "ytick.labelsize": 14})
sns.lineplot(x=tips["total_bill"], y=tips["tip"], linewidth=2, color="red")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=tips)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(tips["total_bill"], label="total_bill")
plt.plot(tips["tip"], label="tip")
plt.plot(tips["size"], label="size")
plt.legend()
plt.show()
```

So we can see using seaborn we can save many lines of code.

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=tips["total_bill"], linewidth=1.5, label="Total Bill", color="blue")
sns.lineplot(data=tips["tip"], linewidth=1.5, label='Tip', color="red")
plt.show()
```

```python
# shoe groups with different colours using hue
plt.figure(figsize=(20,8))
plt.style.use('seaborn-darkgrid')
sns.lineplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
# Color and line dashing to represent 2 different grouping variables using "hue" & "style"

plt.figure(figsize=(20,8))
plt.style.use('seaborn-darkgrid')
sns.lineplot(data=tips, x="total_bill", y="tip", hue="time", style="sex", linewidth=2)
plt.show()
```

```python
# combine lineplot using replot

plt.figure(figsize=(10,6))
plt.style.use('seaborn-darkgrid')
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="time", kind="line", height=6, aspect=1)
plt.show()
```

# Scatter Plot

```python
tips.head()
```

```python
tips.tail()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip)
plt.show()
```

```python
# changing the color of data points using "color" parameter

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, color="green")
plt.show()
```

```python
# chnage the shape of data points using "marker" parameter

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, color="green", marker=">")
plt.show()
```

```python
# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.sex)
plt.show()
```

```python
# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.smoker)
plt.show()
```

```python
# setting X limit using "plt.xlim"
# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
plt.xlim([0,60])
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.sex)
plt.show()
```

```python
# showing two different grouping variables using "hue" and "style" parameter

# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.sex, style=tips.smoker)
plt.show()
```

```python
# varying the size of the points for a quantitative variable using "size" parameter

plt.figure(figsize=(7,7))
# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.sex, size=tips.tip, sizes=(10,50))
plt.show()
```

# Relplot

Relplot provides access to several different axes-level functions like line & scatter. 

```python
sns.set_style("white")
```

```python
# Drawing scatter plot using replot

# show groups with different colors using "hue"

plt.figure(figsize=(6,6))
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", aspect=2)
plt.show()
```

```python
# Use "Pallete" to specify the colors to be used for different levels of the hue

plt.figure(figsize=(6,6))
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", aspect=2, palette=["red", "green"])
plt.gcf().text(.5,.99, "Rel Plot", fontsize=40, color='Black', ha='center', va='center')
plt.show()
```

```python
# Facet along the columns to show a categorical variable using "col" parameter

plt.figure(figsize=(6,6))
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="day", style="sex", kind="line")
plt.show()
```

```python
# Facet along the columns to show a categorical variable using "col" and "row" parameter

plt.figure(figsize=(6,6))
sns.set(rc={"axes.facecolor": "lightyellow"})
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="day", row="sex", kind="line",)
plt.show()
```
