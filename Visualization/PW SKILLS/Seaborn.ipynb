# Seaborn: statistical data visualization

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
## Applying the default theme
sns.set_theme()
```

```python
# load dataset seaborn contains many dataset by default and one of them is iris dataset

iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.shape
```

```python
sns.scatterplot(x = iris.sepal_length, y= iris.sepal_width)
```

```python
sns.scatterplot(x = iris.sepal_length, y= iris.petal_length)
```

```python
sns.displot(iris.sepal_length)
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(data = tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size", )
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip)
```

```python
tips["smoker"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", hue = "smoker", data = tips)
```

```python
sns.relplot(x="total_bill", y="tip", data = tips, style="size", hue="size")
```

```python
sns.relplot(x="total_bill", y="tip", data = tips, style="size", hue="time")
```

### day v/s total_bill

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
tips["day"].value_counts()
```

```python
sns.catplot(y="tip", x="smoker", data=tips)
```

```python
## joint graph

sns.jointplot(x=tips.total_bill, y=tips.tip)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
