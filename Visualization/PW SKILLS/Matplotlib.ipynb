```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Matplotlib

Matplotlib is a comprehensive library for creating static, animated and interactive data visualizations in Python. Matplotlib makes easy things easy and hard things possible.

```python
x = np.linspace(1,10, 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is x label data")
plt.ylabel("This is y label data")
plt.title("This is title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x,y, c=colors, alpha= .5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Random Data")
plt.show()
```

```python
a = ["a", "b", "c", "d", "e"]
b = np.random.randint(1,10, 5)

plt.bar(a,b)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Bar Plot")
plt.show()
```

```python
a = ["a", "b", "c", "d", "e"]
b = np.random.randint(1,10, 5)

plt.barh(a,b)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Bar Plot")
plt.show()
```

```python
x = np.linspace(1,10, 200)
y = np.sin(x)

plt.figure(figsize=(8,2))
plt.xlabel("X label")
plt.ylabel("Y label")
plt.plot(x,y, "--b")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, c='b', alpha= .5, s=100, marker='x')
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Random Data")
plt.show()
```
