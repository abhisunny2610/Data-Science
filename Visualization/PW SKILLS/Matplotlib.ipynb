```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x,y, c=colors, alpha= .5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Random Data")
plt.show()
```

```python
a = ["a", "b", "c", "d", "e"]
b = np.random.randint(1,10, 5)

plt.bar(a,b)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Bar Plot")
plt.show()
```

```python
a = ["a", "b", "c", "d", "e"]
b = np.random.randint(1,10, 5)

plt.barh(a,b)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Bar Plot")
plt.show()
```

```python
x = np.linspace(1,10, 200)
y = np.sin(x)

plt.figure(figsize=(8,2))
plt.xlabel("X label")
plt.ylabel("Y label")
plt.plot(x,y, "--b")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, c='b', alpha= .5, s=100, marker='x')
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Random Data")
plt.show()
```

# pyplot

`matplotlib.pyplot` is a collection of functions that makes matplotlib work like MATLAB. Each `pyplot` function makes some change to a figure: eg., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
plt.plot([1,2,3,4])
plt.ylabel('Some numbers')
plt.show()
```

## Formatting the style of your plot

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. The default format string is 'b-', which is a solid blue line. For example, to plot the above with red circles, you would issue

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis((0,6,0,20))
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

# reddashes, blue squars and green trainagles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
