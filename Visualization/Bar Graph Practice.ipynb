# Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostily written in python, few segments are written in C, Objective-C and Javascript for plotform compatibility.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Create a Bar Plot

```python
data = {"c":20, "c++":15, "python": 50, "java":10, "javascript":20}

df = pd.DataFrame(data=list(data.items()), columns=["Course", "Students"])

fig , axes = plt.subplots(figsize=(4,3))

axes.bar(df.Course, df.Students, color="red", width=0.5)
axes.set_xlabel("Courses")
axes.set_ylabel("Enrolled Students")
plt.title("Students enrolled in different courses")
plt.show()
```

## Multiple Bar Plots

```python
## set bar width 
barWidth = 0.25
fig  = plt.subplots(figsize=(12,8))

## set hight of bars 
IT = [25,30,1,8,22]
ECE=[28,6,16,5,10]
CSE = [29, 3, 24, 25, 17]

## set the position of bar on X axis
br1=np.arange(len(IT))
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2]

plt.bar(br1, IT, color ='r', width = barWidth, 
        edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, 
        edgecolor ='grey', label ='ECE') 
plt.bar(br3, CSE, color ='b', width = barWidth, 
        edgecolor ='grey', label ='CSE')

# Adding Xticks 
plt.xlabel('Branch', fontweight ='bold', fontsize = 15) 
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], 
        ['2015', '2016', '2017', '2018', '2019'])
 
plt.legend()
plt.show()
```

## Stacked bar plot

```python
N = 5
 
boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)   
width = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,
             bottom = boys, yerr = girlStd)
 
plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))
 
plt.show()
```

## Draw horizontal bar chart

for creating horizontal bar chart use `barh()` function instead of `bar(`

```python

import matplotlib.pyplot as plt
 
x=['one', 'two', 'three', 'four', 'five']
 
# giving the values against
# each value at x axis
y=[5, 24, 35, 67, 12] 
plt.bar(x, y)
 
# setting x-label as pen sold
plt.xlabel("pen sold") 
 
# setting y_label as price
plt.ylabel("price")   
plt.title(" Vertical bar graph")
plt.show()
```
