# Imputation Technique

### Mean Imputation (MI)
**Mean Imputation** is one such method in which the mean of the observed values for each avriable is computed and the missing values for that variable are imputed by this mean.This method can lead into severely biased estimates even if data are **MCAR**.


When data is normally distributed this technique in best.

```python
df.head()
```

```python
sns.histplot(df["Schooling"], kde=True, bins=30)
```

```python
df["Schooling"].isnull().sum()
```

```python
df["Schooling Mean"]= df["Schooling"].fillna(df['Schooling'].mean())
```

```python
df[["Schooling Mean", "Schooling"]]
```

```python
df["Schooling Mean"].isna().sum()
```

### Median Imputation

If dataset have an outliers we generally use median imputation technique to fill the null values.

```python
df["Schooling Median"] = df["Schooling"].fillna(df["Schooling"].median())
```

```python
df[["Schooling Median","Schooling Mean", "Schooling"]]
```

```python
df[df["Schooling"].isnull()]
```
